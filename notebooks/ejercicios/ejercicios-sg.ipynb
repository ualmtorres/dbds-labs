# Ejercicios sobre la base de datos Sporting Goods

## Configuración del entorno

Asegurarse de seguir estos pasos para configurar el entorno de trabajo:

* Instalar los paquetes necesarios para trabajar con bases de datos en Jupyter Notebook (si no están ya instalados).


* Importar la extensión SQL en Jupyter Notebook para ejecutar consultas SQL directamente desde las celdas del notebook.


* Cargar la extensión SQL y conectarse a la base de datos Sporting Goods utilizando la URI de conexión a la instalación de MySQL utilizada.

* Ejecutar el [script de inicialización de la base de datos](https://raw.githubusercontent.com/ualmtorres/dbds-tutorial/refs/heads/gh-pages/SG/SG.sql) para crear las tablas y cargar los datos iniciales en la base de datos Sporting Goods. Para más información, consultar el [esquema de base de datos Sporting Goods](https://ualmtorres.github.io/dbds-tutorial/models/SG.pdf).

### Listar productos (muestra)

Enunciado: Mostrar id y nombre de productos ordenados por nombre (limitado a 10 resultados).

### Contar productos

Enunciado: Obtener el número total de productos en la tabla `s_product`.

### Inventario en almacén 101

Enunciado: Mostrar los siguientes datos de los 10 primeros productos del almacén `101`: Mostrar `product_id`, `amount_in_stock` y `reorder_point`.

### Stock por almacén y producto

Enunciado: Mostrar `product_id`, nombre del producto, `warehouse_id` y `amount_in_stock` para los productos cuyo nombre contiene el texto `Prostar` y que están en los almacenes `301` o `401`. Ordenar por `warehouse_id` y `product_id`. Limitar a 10 resultados.

### Stock total por almacén

Enunciado: Calcular el stock total disponible por cada almacén (suma de `amount_in_stock`). Mostrar `warehouse_id` y `total_stock` y ordenar por `total_stock` de forma descendente.

### Ventas totales por producto (importe)

Enunciado: Calcular las ventas totales por producto como SUM(quantity * price) usando las tablas `s_item` y `s_ord` (suponiendo que todas las filas en `s_item` corresponden a pedidos facturados). Mostrar `product_id`, nombre del producto y `ventas_totales`.  Presentar en orden descendente de ventas y limitar a 10 resultados.

### Cantidad vendida por producto

Enunciado: Calcular la cantidad total vendida por producto (SUM(`quantity`)). Mostrar `product_id`, nombre del producto y `cantidad_total_vendida`. Ordenar por cantidad en orden descendente y limitar a 10 resultados.

### Ventas por región

Enunciado: Calcular las ventas totales de pedidos por nombre de región. Mostrar `region_name` y `ventas_region`, ordenado por `ventas_region` descendente.

### Stock total por producto

Enunciado: Obtener, para cada producto, el stock total sumando existencias en inventario. Mostrar `product_id`, nombre del producto y total de stock de aquellos productos con stock mayor que 2000 unidades. Ordenar por stock descendente.

### Productos sin stock en algún almacén

Enunciado: Listar los productos que aparecen con cantidad 0 en inventario (mostrar `product_id`, `warehouse_id` y `cantidad`).

### Clientes por región

Enunciado: Mostrar el número de clientes por región. Mostrar `region` y `num_clientes`, ordenado por `num_clientes` en orden descendente.

### Pedidos de gran importe

Enunciado: Listar los pedidos cuyo total sea superior a 100000. Mostrar `id`, `customer_id`, `date_ordered` y `total`. La ordenación será por `total` descendente y la fecha se mostrará en formato 'DD-MM-YYYY'.

### Top vendedores por ventas totales

Enunciado: Calcular las ventas totales en pedidos gestionads por cada representante de ventas (sales_rep). Mostrar `id` del empleado, nombre (`first_name` + `last_name`) y suma de totales. Mostrar en orden descendente de suma de totales y mostrar top 5.

### Inventario que necesita reposición

Enunciado: Listar las líneas de inventario cuyo `amount_in_stock` sea menor o igual al `reorder_point` (o con `amount_in_stock` `NULL`). Mostrar `product_id`, `warehouse_id`, `amount_in_stock` y `reorder_point`.
